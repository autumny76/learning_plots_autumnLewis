```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
# group by genre and calculate the average total streams
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.savefig("../plots/AverageTotalStreamsbyGenre.png")
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.Figure(figsize=(10,6))
colors = ["#7b92c9"]*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#f030ad"


plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.savefig("../plots/AverageTotalStreamsbyArtist.png")
plt.show()
```

```python
plt.figure(figsize=(12,8))

special_artist = "Ariana Grande"
colors = ["#f030ad" if artist == special_artist else "gray" for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.9)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Ariana Grande Ranks 2nd in Total Average Streams")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.savefig("../plots/ArianaGrandeRanks2ndinTotalAvgStreams.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"],
    spotify_df["Total Streams (Millions)"],
    color = "#08cc15",
    alpha = 0.8,
    s= 15)

plt.xlabel("Streams this month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.savefig("../plots/MonthlyListensvsTotalStreams.png")
plt.show()
```

```python
plt.figure(figsize=(10,6))

featured_genre = "Pop"

colors = ['#08cc15' if genre == featured_genre else "#d1d5d1" for genre in spotify_df["Genre"]]

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"],
    spotify_df["Total Streams (Millions)"],
    color = colors,
    alpha = 0.8,
    s= 15)

plt.xlabel("Streams this month")
plt.ylabel("Total Streams")
# plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Montly Listens or Total Streams?", fontsize = 18, pad=35)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```

```python
# Line Chart

yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(12,6))
plt.plot(yearly_avg.index, yearly_avg.values)
plt.plot(
    yearly_avg.index,
    yearly_avg.values,
    marker="+",
    linestyle = "--",
    color = "#20a32b",
    linewidth = 0.5)

plt.xlabel("Release Year")
plt.ylabel("Average Total Streams")
plt.title("Did Listeners Stream Less in 2022?", fontsize = 18, pad=20)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axvline(
    2022,
    color= "#063b0a",
    linewidth = 0.5,
    alpha = 0.7)

ax.text(
    2022+0.05,
    2225,
    "Dip In Total Streams",
    color= "#063b0a",
    fontsize=8
)

plt.ylim(0, 3500)

plt.tight_layout()
plt.savefig("../plots/LineChartLessStreamsin2022.png")
plt.show()
```

```python
# CREATING PIE CHARTS
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    platform_totals,
    labels=platform_totals.index,
    startangle=110,
    colors=["gray", "#1aaa10"],
    autopct="%1.1f%%")

plt.title("Near Even Split Between Premium and Free Users")

plt.legend(
    labels = platform_totals.index,
    loc="lower right",
    title = "Platform"
)

plt.show()
```

```python
#another option to do pie chart
fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct="%1.1f%%",
    startangle=110,
    colors=["gray", "#1aaa10"],)

ax.set_title(
    "Near Even Split Between Premium and Free Users",
    fontsize=20,
    pad=15,
    loc="center"
)

for autotext in autotexts:
    autotext.set_fontsize(25)
    autotext.set_color("black")
    autotext.set_weight("bold")

ax.legend(
    wedges,
    platform_totals.index,
    loc="lower center",
    bbox_to_anchor = (0.5, 0.95),
    ncol= 2
)
plt.tight_layout()
plt.savefig("../plots/PiePremiumVSFreeUsers.png")
plt.show()
```

```python
monthly_streams = spotify_df.groupby("Country")["Streams Last 30 Days (Millions)"].mean().sort_values(ascending=False)

# Create a color list
colors = ["purple"] * len(monthly_streams)

# # Highlight the max
# max_index = monthly_streams.idxmax()
# highlight_position = monthly_streams.index.get_loc(max_index)
# colors[highlight_position] = "#f030ad"

# Plot using bar chart
plt.figure(figsize=(12,6))
plt.bar(monthly_streams.index, monthly_streams.values, color=colors, edgecolor="white")

plt.xlabel("Country")
plt.ylabel("Streams Last 30 Days (Millions)")
plt.title("Streams of Last 30 Days by Country", fontsize=16, pad=20)
plt.xticks(rotation=45, ha="right")

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
# Box Plots

plt.boxplot(
    [spotify_df[spotify_df["Platform Type"]=="Free"]["Skip Rate (%)"],
    spotify_df[spotify_df["Platform Type"]=="Premium"]["Skip Rate (%)"]]
)
plt.xticks([1,2],["Free", "Premium"])
plt.xlabel("Platform Type")
plt.ylabel("Skip Rate (%)")
plt.title("Skip Rate (%) by Platform Type", fontsize=14, pad=15)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
# Heatmaps
numberic_spotify_df = spotify_df.select_dtypes(include=[np.number])
corr = numeric_spotify_df.corr()

im = plt.imshow(corr, vmin=-1, vmax=1, cmap="coolwarm")
plt.colorbar(im, label="Pearson Corellation")

plt.xticks(range(len(corr.columns)), corr.columns, rotation= 45, ha="right")
plt.yticks(range(len(corr.columns)), corr.columns)


plt.show()
```

```python
# Count Plots

genre_counts = spotify_df["Genre"].value_counts().sort_values(ascending=False)

colors = ["#B57EDC"] * len(genre_counts)  # lavender hex code

plt.figure(figsize=(10,6))
plt.bar(genre_counts.index, genre_counts.values, color=colors)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Genre Counts")
plt.title("Counts by Genre")
plt.xticks(rotation=45, ha="right")

plt.savefig("../plots/CountsbyGenre.png")
plt.show()
```
